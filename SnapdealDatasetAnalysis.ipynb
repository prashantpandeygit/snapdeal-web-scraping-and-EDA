```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/CSVs/Snapdeal WebScraped Dataset.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

```python
def get_short_product_name(product_name):
    words = product_name.split()[:3]
    return ' '.join(words)
df['Short Product Name'] = df['Product Name'].apply(get_short_product_name)
```

```python
df.set_index('Short Product Name', inplace=True)
```

```python
Q1 = df['Product Rating Count'].quantile(0.25)
Q3 = df['Product Rating Count'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df.drop(df[(df['Product Rating Count'] < lower_bound) | (df['Product Rating Count'] > upper_bound)].index, inplace=True)
```
