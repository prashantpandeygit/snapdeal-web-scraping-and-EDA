```python
url = 'https://www.snapdeal.com/search?keyword=earphones%20under%201000&santizedKeyword=mobiles+under+10000&catId=0&categoryId=0&suggested=false&vertical=p&noOfResults=20&searchState=&clickSrc=go_header&lastKeyword=&prodCatId=&changeBackToAll=false&foundInAll=false&categoryIdSearched=&cityPageUrl=&categoryUrl=&url=&utmContent=&dealDetail=&sort=rlvncy'
```

```python
r=requests.get(url)
r
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
soup
```

```python
soup.prettify()
```

```python
product_names = []
prices = []
description = []
product_rating_count = []
color = []
discount_percent = []
maximum_price = []
below_600 = []
```

```python
product_names = [tag['title'] for tag in soup.find_all('p', class_='product-title')]
```

```python
product_names
```

```python
len(product_names)
```

```python
prices = [tag['data-price'] for tag in soup.find_all('span', class_='lfloat product-price')]
```

```python
prices
```

```python
prices = [int(price) for price in prices]
```

```python
len(prices)
```

```python
description = [tag['title'] for tag in soup.find_all('p', class_='product-title')]
```

```python
description
```

```python
len(description)
```

```python
product_rating_count = soup.find_all('p', class_='product-rating-count')
product_rating_count
```

```python
numbers = []
for tag in product_rating_count:
    text_content = tag.get_text()
    number = ''.join(filter(str.isdigit, text_content))
    numbers.append(int(number))
    product_rating_count = numbers
```

```python
product_rating_count
```

```python
for i in product_rating_count:
    print(type(i))
```

```python
len(product_rating_count)
```

```python
discount_percent = soup.find_all('div', class_='product-discount')
```

```python
discounts = []
for discount in discount_percent:
    text_content_numeric = ''.join(filter(str.isdigit, str(discount)))
    # Convert the resulting string back to an integer
    discount_value = int(text_content_numeric)
    discounts.append(discount_value)
discount_percent = discounts
```

```python
discount_percent
```

```python
for i in discount_percent:
    print(type(i))
```

```python
len(discount_percent)
```

```python
maximum_prices = soup.find_all('span', class_='lfloat product-desc-price strike')
maximum_prices
```

```python
maximum_price = []
for tag in maximum_prices:
    text_content = tag.get_text()
    numeric_value = int(''.join(filter(str.isdigit, text_content.replace(',', ''))))
    maximum_price.append(numeric_value)
```

```python
maximum_price
```

```python
for i in maximum_price:
    print(type(i))
```

```python
len(maximum_price)
```

```python
below_600 = ['Yes' if price < 600 else 'No' for price in prices]
```

```python
below_600
```

```python
df = pd.DataFrame({'Product Name' : product_names, 'Maximum Price' : maximum_price, ' Sale Price' : prices, 'Product Rating Count' : product_rating_count, 'Discount Percent' : discount_percent, 'Description' : description,  'Below 600' : below_600 })
```
